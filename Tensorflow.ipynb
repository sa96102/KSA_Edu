```
import tensorflow as tf
import numpy as np

# initialize two constants
m1 = np.matrix('1 2; 3 4', dtype='int32')
tm1 = tf.constant(m1)

m2 = np.matrix('10 20; 30 40', dtype='int32')
tm2 = tf.constant(m2)
print('numpy matrices \nm1:' + str(m1) + '\nm2:' + str(m2))
print('tf matrices \nm1:' + str(tm1) + '\nm2:' + str(tm2))

#multiplication
npResult = np.matmul(m1, m2)
print('numpy result\nm1xm2:' + str(npResult))

tfResult = tf.matmul(tm1, tm2)
print('tf result\nm1xnm2:' + str(tfResult))

sess = tf.Session()
tfOut = sess.run(tfResult) # tfOut의 값은 numpy matrix.
print('--------------------------------------------')
print(tfOut)

sess.close()
```

```
import tensorflow as tf
import numpy as np
from tensorboardcolab import *

# Tensorboard 를 Colab에서 사용하기 위한 코드
tbc=TensorBoardColab()

# Graph, Loss 변화 등을 기록하기 위한 Writer
writer = tbc.get_writer()

# Tensorflow의 모든 Graph를 초기화 
tf.reset_default_graph()

# Session 객체 생성
with tf.Session() as sess:
  
  # Matrix 객체 생성
  m1 = np.matrix('1 2;3 4', dtype=np.int32)
  tm1 = tf.constant(m1)

  m2 = np.matrix('10 20;30 40 ', dtype=np.int32)
  tm2 = tf.constant(m2)

  print('numpy matrices \n' + str(m1) + '\n' + str(m2))
  print('tensorflow matrices \n' + str(tm1) + '\n' + str(tm2))

  # 행렬 곱
  npr = np.matmul(m1, m2)
  print('numpy result \n' + str(npr))

  # Tensorflow 행렬 곱 그래프 노드 추가
  tfr = tf.matmul(tm1, tm2)
  print('tensorflow result \n' + str(tfr))

  # Writer에 만든 Graph 기록
  writer.add_graph(sess.graph)
  
  # Session 실행
  tfOut = sess.run(tfr)
  
  # 실행 결과 출력
  print(tfOut)
  
  # Session 닫기
  sess.close()

 # Writer 닫기
writer.close()
```
