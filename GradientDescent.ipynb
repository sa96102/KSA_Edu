```
import numpy as np
import matplotlib.pyplot as plt

def j(x):
    return np.power(x, 2)

def dj(x):
    return 2*x

def gd(initX, lr, precision, maxIter):
    x = initX
    curX = x
    sx = []
    sj = []
    for i in range(maxIter):
        x = curX - lr * dj(curX)
        if abs(x-curX) < precision:
            return curX, sx, sj
        curX = x
        sx.append(x)
        sj.append(dj(x))
        #print(curX)
    return curX, sx, sj
    

def main():
    x = np.arange(-6, 6, 1)
    print(x)
    print(j(x))
    
    plt.plot(x, j(x))
    plt.plot(x, dj(x))
    plt.grid()
    plt.show()
    
    lr = 0.1 # learning rate
    precision = 0.001 #threshold, stopping criterion
    maxIter = 100 # number of iterations
    optX, sx, sy = gd(4, lr, precision, maxIter)
    print(optX) # = 0.004951760157141522
    plt.plot(sx, sy)
    plt.grid()
    plt.show()

    
if __name__ == "__main__":
    main()
```
